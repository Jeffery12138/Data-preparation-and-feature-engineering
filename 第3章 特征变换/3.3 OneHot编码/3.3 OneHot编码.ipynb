# 项目案例

```python
df = pd.DataFrame({
    "color": ['green', 'red', 'blue', 'red'],
    "size": ['M', 'L', 'XL', 'L'], 
    "price": [29.9, 69.9, 99.9, 59.9],
    "classlabel": ['class1', 'class2', 'class1', 'class1']
})
df
```

```python
size_mapping = {'XL': 3, 'L': 2, 'M': 1}
df['size'] = df['size'].map(size_mapping)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
fs = ohe.fit_transform(df[['color']])
fs_one = pd.DataFrame(fs.toarray()[:, 1:],
                     columns=['color_green',
                             'color_red'])
df = pd.concat([df, fs_one], axis=1)
df
```

# 动手练习

```python
# 第一题
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
fc = ohe.fit_transform(df[['classlabel']])
fc_one = pd.DataFrame(fc.toarray(),
                     columns=['classlabel_class1',
                             'classlabel_class2'])
df = pd.concat([df, fc_one], axis=1)
df
```

```python
# 第二题
df = pd.read_csv("./data/breast-cancer.data", header=None).iloc[:, 1:]
df
```

```python
dataset = df.values
X = dataset[:, 0:8]
X = X.astype(str)
Y = dataset[:, 8:]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
encoded_x = None
for i in range(0, X.shape[1]):
    label_encoder = LabelEncoder()  # 数值化
    feature = label_encoder.fit_transform(X[:, i])
    feature = feature.reshape(X.shape[0], 1)
    onehot_encoder = OneHotEncoder(sparse=False)  # OneHot编码
    feature = onehot_encoder.fit_transform(feature)
    if encoded_x is None:
        encoded_x = feature
    else:
        encoded_x = np.concatenate((encoded_x, feature), axis=1)
print("X shape: ", encoded_x.shape)
```

```python
label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(Y)
label_encoder_y = label_encoder.transform(Y)
label_encoder_y
```
