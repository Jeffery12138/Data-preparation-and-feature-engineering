```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text)

def extract_text(soup, id_value):
    txt = soup.find("div", {"id": id_value})
    title = txt.h2.text
    content = txt.p.text
    return {title: content}

result = {}
for id_value in {"stack-overflow",
                "send-a-tweet",
                "file-an-issue",
                "e-mail",
                "irc"}:
    web_txt = extract_text(soup, id_value)
    result.update(web_txt)
    
import pprint
pprint.pprint(result)
```

# 项目案例

(https://movie.douban.com/coming)

```python
import requests
from bs4 import BeautifulSoup

url = "https://movie.douban.com/coming"
response = requests.get(url)
response
```

```python
response.status_code
```

```python
response.text
```

```python
# 添加headers信息
headers={
    'User-Agent': 
    'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'}
response = requests.get(url, headers=headers)
response
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
table = soup.find("table", attrs={"class": "coming_list"})
table
```

```python
table_body = table.find("tbody")
table_body
```

```python
rows = table_body.find_all("tr")
rows
```

```python
rows[0]
```

```python
data = []
for row in rows:
    cols = row.find_all("td")
    cols = [ele.text.strip() for ele in cols]
    data.append(cols)
data
```

```python
data = []
for row in rows:
    id_value = row.find("a")['href'][-9:-1].replace("/", "")
    cols = row.find_all("td")
    cols = [ele.text.strip() for ele in cols]
    cols.append(id_value)
    data.append(cols)
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
df.columns = ['上映日期', '片名', '类型', '制片国家/地区', '想看', 'ID']
df
```

```python
# 获取影片详情
url_fore = 'https://movie.douban.com/subject/'

def movie(id_value):
    url = url_fore + id_value
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'lxml')
    
    movie_infos = soup.find("div", attrs={"id": "info"})
    directors = movie_infos.find_all(rel="v:directedBy")
    dlst = [d.text for d in directors if d]
    actors = movie_infos.find_all(rel="v:starring")
    alst = [actor.text for actor in actors if actor]
    
    director_str = '|'.join(dlst)
    actor_str = "|".join(alst)
    
    return [director_str, actor_str, id_value]
```

```python
movies_lst = []
for i in df['ID']:
    infos = movie(i)
    movies_lst.append(infos)
movies_lst
```

```python
movies_df = pd.DataFrame(movies_lst, columns=['导演', '主演', 'ID'])
movies_df
```

```python
datas = pd.merge(df, movies_df, on='ID')
datas
```

```python
datas.to_csv("./movies.csv")
```
