# 项目案例

```python
cpi = pd.read_excel(path + "/cpi.xls")
cpi.head(4)
```

```python
cpi
```

```python
cpi.columns = cpi.iloc[1]
cpi = cpi[2:]
cpi
```

```python
cpi.drop([11, 12], axis=0, inplace=True)
cpi
```

```python
cpi['cpi_index'] = ['总体消费', '食品烟酒', '衣着', '居住', '生活服务', '交通通信', '教育娱乐', '医保', '其他']
cpi
```

```python
cpi.drop(['指标'], axis=1, inplace=True)
cpi
```

```python
cpi.reset_index(drop=True, inplace=True)
cpi
```

```python
# 将列索引的名字设置为空
cpi.columns.rename("", inplace=True)
cpi
```

```python
cpi.info()
```

```python
for column in cpi.columns[:-1]:
    cpi[column] = pd.to_numeric(cpi[column])
cpi.dtypes
```

```python
%matplotlib
import matplotlib.pyplot as plt
plt.bar(cpi.iloc[5, :-1].index, cpi.iloc[5, :-1].values)
plt.grid()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(cpi.iloc[5, :-1].index, cpi.iloc[5, :-1].values)
plt.grid()
```

```python
%matplotlib
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.font_manager import FontProperties

myfont = FontProperties(fname='fonts/msyh.ttc')
matplotlib.rcParams['axes.unicode_minus'] = False

plt.bar(cpi.iloc[5, :-1].index, cpi.iloc[5, :-1].values, width=0.6)
plt.xticks(rotation=45, fontsize=10, fontproperties=myfont)
plt.grid()
```

```python
plt_book = imgplt.imread("./work/images/book.png")
plt_book.shape
```

```python
pd.concat?
```

```python
part_img = plt_img[100:200, 200:360, :]
part_img.shape
```

```python
part_book = plt_book[200:240, 300:460,:]
part_book.shape
```

```python
com_img = pd.concat([part_img, part_book])
plt.imshow(com_img)
```

```python
import numpy as np
com_img = np.vstack([part_img, part_book])
plt.imshow(com_img)
```

```python
import numpy as np
from PIL import Image
import cv2
img1 = cv2.imread('work/images/laoqi.png', -1)
img2 = cv2.imread("work/images/book.png", -1)
part1 = img1[50:260, 100:280]
part2 = img2[300:, 100:280]
new_img = np.vstack((part1, part2))
Image.fromarray(new_img)
```
