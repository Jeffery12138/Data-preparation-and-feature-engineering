```python
# (1) 读取数据，并随机抽样，制作缺失值
persons = pd.read_csv("./data/Person.csv")
pdf = persons.sample(20)  # 随机抽取20个样本
pdf['Height-na'] = np.where(pdf['Height'] % 5 == 0, np.nan, pdf['Height'])
pdf
```

```python
# (2) 用平均值填补缺失数据
pdf['Height-na'].fillna(pdf['Height-na'].mean(), inplace=True)
pdf
```

```python
pdf['Height'].describe()
```

```python
pdf['Height-na'].describe()
```

# 动手练习

```python
pdf = persons.sample(20)
pdf['Height-na'] = np.where(pdf['Height'] % 5 == 0, np.nan, pdf['Height'])
pdf
```

```python
pdf['Height-na'].fillna(pdf['Height'].median(), inplace=True)
pdf
```

```python
pdf['Height'].describe()
```

```python
pdf['Height-na'].describe()
```

```python
pdf = persons.sample(20)
pdf['Height-na'] = np.where(pdf['Height'] % 5 == 0, np.nan, pdf['Height'])
pdf
```

```python
pdf['Height-na'].fillna(pdf['Height'].mode(), inplace=True)
pdf
```

```python
pdf.mode()
```

```python
pdf['Height-na'].mode()[0]
```

```python
type(pdf['Height-na'].mode())
```

```python
pdf['Height-na'].fillna(pdf['Height-na'].mode()[0], inplace=True)
pdf
```

```python
pdf['Height'].describe()
```
