```python
# (1) 读取数据，并随机抽样，制作缺失值
persons = pd.read_csv("./data/Person.csv")
pdf = persons.sample(20)  # 随机抽取20个样本
pdf['Height-na'] = np.where(pdf['Height'] % 5 == 0, np.nan, pdf['Height'])
pdf
```

```python
# (2) 用平均值填补缺失数据
pdf['Height-na'].fillna(pdf['Height-na'].mean(), inplace=True)
pdf
```

```python
pdf['Height'].describe()
```

```python
pdf['Height-na'].describe()
```

# 扩展探究

```python
pdf2 = persons.sample(20)
pdf2['Height-na'] = np.where(pdf2['Height'] % 5 == 0, np.nan, pdf2['Height'])

from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
col_values = imp_mean.fit_transform(pdf2['Height-na'].values.reshape((-1, 1)))
col_values
```
